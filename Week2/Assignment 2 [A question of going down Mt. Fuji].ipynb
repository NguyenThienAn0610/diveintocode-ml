```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
csv_path = "mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
point_num = fuji[:,0]
latitude = fuji[:,1]
longitude = fuji[:,2]
elevation = fuji[:,3]
distance_from_point_0 = fuji[:,4]
```

# Problem 1

```python
plt.xlabel("Position")
plt.ylabel("Elevation")
plt.title("Mt. Fuji")
plt.plot(point_num, elevation)
plt.show()
```

# Problem 2

```python
def cal_slope_1_point(x, y, pos):
    if pos > 0 or pos < len(x):
        gradient = (y[pos] - y[pos - 1]) / (x[pos] - x[pos - 1])
    else:
        gradient = 0
    return gradient
pos = 136
print("Slope at point", pos, "is", cal_slope_1_point(point_num, elevation, pos))
```

# Problem 3

```python
def cal_next_destination(current_location, current_gradient, alpha = 0.2):
    destination = int(round(current_location - alpha * current_gradient))
    if destination < 0 or destination > 300:
        return -1
    else:
        return destination
print("Next destination of position", pos, "is", cal_next_destination(pos, cal_slope_1_point(point_num, elevation, pos), 0.2))
```

# Problem 4

```python
def goes_down_mountain(pos, alpha = 0.2):
    move_record = []
    next_destination = cal_next_destination(pos, cal_slope_1_point(point_num, elevation, pos), alpha)
    for _ in range(1000):
        move_record.append(next_destination)
        next_destination = cal_next_destination(next_destination, cal_slope_1_point(point_num, elevation, next_destination), alpha)
    return move_record
print(goes_down_mountain(136))
```

# Problem 5

```python
plt.xlabel("move time")
plt.ylabel("position")
plt.title("Descending process")
plt.plot(point_num, elevation)
plt.plot(goes_down_mountain(136), [elevation[i] for i in goes_down_mountain(136)],'r.')
plt.show()
```

# Problem 6

```python
plt.xlabel("move time")
plt.ylabel("position")
plt.title("Descending process")
plt.plot(point_num, elevation)
plt.plot(goes_down_mountain(142), [elevation[i] for i in goes_down_mountain(142)],'r.')
plt.show()
```
