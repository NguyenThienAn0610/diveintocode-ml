```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Problem 1

```python
x_ndarray = np.arange(-50, 50.1, 0.1)
x_ndarray = x_ndarray.reshape((1001,1))
y_ndarray = 1/2 * x_ndarray + 1
print(x_ndarray, "\n" ,y_ndarray)
print(x_ndarray.shape)
```

# Problem 2

```python
combined_array = np.hstack((x_ndarray, y_ndarray))
print(combined_array)
print(combined_array.shape)
```

# Problem 3

```python
d_a = (combined_array.T[1][1:] - combined_array.T[1][:-1]) / (combined_array.T[0][1:] - combined_array.T[0][:-1])
print(d_a)
```

# Problem 4

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of the linear function")
plt.plot(x_ndarray, y_ndarray)
plt.show()
plt.xlabel("x")
plt.ylabel("Gradient")
plt.title("Graph of the gradient")
plt.plot(x_ndarray[1:], d_a)
plt.show()
```

# Problem 5

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    array_x = np.arange(x_range[0], x_range[1], x_range[2])
    print("array_x shape:", array_x.shape)
    array_y = function(array_x)
    print("array_y shape:", array_y.shape)
    array_xy = np.hstack((array_x.reshape((len(array_x), 1)), array_y.reshape((len(array_y), 1))))
    print("array_xy shape:", array_xy.shape)
    gradient = (array_y[1:] - array_y[:-1]) / (array_x[1:] - array_x[:-1])
    print("gradient shape:", gradient.shape)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Graph of the linear function")
    plt.plot(array_x, array_y)
    plt.show()
    plt.xlabel("x")
    plt.ylabel("Gradient")
    plt.title("Graph of the gradient")
    plt.plot(array_x[1:], gradient)
    plt.show()
    return array_xy, gradient
def function1(array_x): # y = x^2
    array_y = np.power(array_x, 2)
    return array_y
def function2(array_x): # y = 2x^2 + 2^x
    array_y = 2 * np.power(array_x, 2) + np.power(2, array_x)
    return array_y
def function3(array_x): # y = sin(x^(1/2))
    array_y = np.sin(np.power(array_x, 1/2))
    return array_y
array_xy_1, gradient_1 = compute_gradient(function1, (-50, 50.1, 0.1))
array_xy_2, gradient_2 = compute_gradient(function2, (-50, 50.1, 0.1))
array_xy_3, gradient_3 = compute_gradient(function3, (0, 50.1, 0.1))
```

# Problem 6

```python
min_y = np.min(array_xy_1.T[1]), np.min(array_xy_2.T[1]), np.min(array_xy_3.T[1])
print(min_y)
idx_min_y = np.argmin(array_xy_1.T[1]), np.ndarray.argmin(array_xy_2.T[1]), np.ndarray.argmin(array_xy_3.T[1])
print(idx_min_y)
print("Slope before function 1:", gradient_1[idx_min_y[0] - 1])
print("Slope after function 1:", gradient_1[idx_min_y[0] + 1])
print("Slope before function 2:", gradient_2[idx_min_y[1] - 1])
print("Slope after function 2:", gradient_2[idx_min_y[1] + 1])
print("Slope before function 3:", gradient_3[idx_min_y[2] - 1])
print("Slope after function 3:", gradient_3[idx_min_y[2] + 1])
```
