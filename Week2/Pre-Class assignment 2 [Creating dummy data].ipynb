```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Problem 1

```python
np.random.seed(0)
mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]]
dataProb1 = np.random.multivariate_normal(mean = mean, cov = cov, size = 500)
print(dataProb1.shape)
print(dataProb1)
```

# Problem 2

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("Problem 2 plot")
plt.scatter(dataProb1.T[0], dataProb1.T[1])
plt.show()
```

# Problem 3

```python
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("Histogram of x1")
plt.xlim(-5, 3)
plt.hist(dataProb1.T[0], bins = 50)
plt.show()
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("Histogram of x2")
plt.xlim(-5, 3)
plt.hist(dataProb1.T[1], bins = 50)
plt.show()
```

# Problem 4

```python
mean_prob_4 = [0, -3]
cov_prob_4 = [[1, 0.8], [0.8, 1]]
dataProb4 = np.random.multivariate_normal(mean = mean_prob_4, cov = cov_prob_4, size = 500)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Problem 4 plot")
plt.scatter(dataProb1.T[0], dataProb1.T[1])
plt.scatter(dataProb4.T[0], dataProb4.T[1])
plt.show()
```

# Problem 5

```python
data_combination_ndarray = np.concatenate((dataProb1 ,dataProb4), axis = 0)
print(data_combination_ndarray.shape)
print(data_combination_ndarray)
```

# Problem 6

```python
label_array_1 = np.zeros((500, 1))
label_array_4 = np.ones((500, 1))
labeled_data_1 = np.hstack((dataProb1, label_array_1))
labeled_data_4 = np.hstack((dataProb4, label_array_4))
labeled_array = np.vstack((labeled_data_1, labeled_data_4))
print(labeled_array)
print(labeled_array.shape)
```
