```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Problem 1

```python
dataset = datasets.load_iris()
print(dataset)
```

```python
X = pd.DataFrame(dataset.data, columns = dataset.feature_names)
print(X)
```

```python
Y = pd.DataFrame(dataset.target, columns = ["Species"])
print(Y)
```

# Problem 2

```python
df = pd.concat([X, Y], axis = 1)
print(df)
```

# Problem 5

```python
# Extract sepal_width in three different ways
# 1st way
print(df[df.columns[1]])

# 2nd way
print(df["sepal width (cm)"])

# 3rd way
print(df.T.iloc[1])
```

```python
# Extract the 50th to 99th data
print(df.loc[50:99])
```

```python
# Extract the 50th to 99th data of the petal_length column
print(df.iloc[50:99]["petal length (cm)"])
```

```python
# Extract data with a petal_width value of 0.2
df.loc[df['petal width (cm)'] == 0.2]
```

### Difference between loc and iloc
<blockquote>
    <ul>
        <li><b>loc</b> is label_based, which means that you have to specify rows and columns based on their rows and columns based on their wor and column labels.</li>
        <li><b>iloc</b> is integer index based, so you have to specify rows and columns by their integer index</li>
    </ul>
</blockquote>

```python
# Box plot
sns.set_theme(style = "whitegrid")
_, axes = plt.subplots(3, 1, figsize = (10,20))
box_plot_setosa = sns.boxplot(ax = axes[0], data = df[df.columns[0:4]][:50]).set_title("setosa")
box_plot_virgicolor = sns.boxplot(ax = axes[1], data = df[df.columns[0:4]][50:100]).set_title("virgicolor")
box_plot_virginica = sns.boxplot(ax = axes[2], data = df[df.columns[0:4]][100:]).set_title("virginica")
```

```python
# Violin plot
_, axes = plt.subplots(3, 1, figsize = (10,20))
violin_plot_setosa = sns.violinplot(ax = axes[0], data = df[df.columns[0:4]][:50]).set_title("setosa")
violin_plot_virgicolor = sns.violinplot(ax = axes[1], data = df[df.columns[0:4]][50:100]).set_title("virgicolor")
violin_plot_virginica = sns.violinplot(ax = axes[2], data = df[df.columns[0:4]][100:]).set_title("virginica")
```

### Difference between box plot and violin plot
<blockquote>
    <ul>
        <li><b>box plot</b> could show the number of observations in the groups using the var width while the violin plot couldn't</li>
        <li><b>violin plot</b> shows a better view of the distribution of data.</li>
    </ul>
</blockquote>
